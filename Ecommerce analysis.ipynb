```python
# E-Commerce Profitability Calculator with Scenario Analysis

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define file path for saving the dataset
file_path = r"C:\Users\admin\Downloads\DT_Ecommerce_Growth_Hacker_Assignment.xlsx"

# Define input parameters
selling_price = 30  # Selling Price per unit
cogs = 15  # Cost of Goods Sold per unit
amazon_fee_percent = 15  # Amazon Fee %
tax_rate = 20  # Tax Rate %
fixed_costs = 10000  # Monthly Fixed Corporate Costs
interest = 1500  # Monthly Bank Loan Interest
units_sold = 5000  # Target monthly sales volume

# Create a dataset with 5000 rows (each row representing one unit sold)
data = []
for i in range(units_sold):
    cac = np.random.uniform(0, 10)  # Random CAC value between 0 and 10
    gross_margin = selling_price - cogs
    amazon_fee = selling_price * (amazon_fee_percent / 100)
    net_margin_per_unit = gross_margin - amazon_fee - cac
    data.append([selling_price, cogs, gross_margin, amazon_fee, cac, net_margin_per_unit])

# Convert data to DataFrame
df = pd.DataFrame(data, columns=["Selling Price", "COGS", "Gross Margin", "Amazon Fee", "CAC", "Net Margin per Unit"])

# Calculate total profit metrics
df["Total Profit Before Tax"] = df["Net Margin per Unit"] * units_sold - fixed_costs - interest
df["Profit After Tax"] = df["Total Profit Before Tax"] * (1 - tax_rate / 100)

# Save the dataset for further analysis in Google Sheets
df.to_excel(file_path, sheet_name="Task1", index=False)

print("Profitability analysis completed and saved successfully!")

# Scenario Analysis
scenarios = [
    {"Units Sold": 5000, "Fixed Costs": 10000},
    {"Units Sold": 4000, "Fixed Costs": 10000},  # Sales drop to 4000
    {"Units Sold": 5000, "Fixed Costs": 15000}   # Fixed costs increase by $5000
]

scenario_results = []

for scenario in scenarios:
    units_sold = scenario["Units Sold"]
    fixed_costs = scenario["Fixed Costs"]
    df["Total Profit Before Tax"] = df["Net Margin per Unit"] * units_sold - fixed_costs - interest
    df["Profit After Tax"] = df["Total Profit Before Tax"] * (1 - tax_rate / 100)
    max_profitable_cac = df[df["Profit After Tax"] >= 0]["CAC"].max()
    scenario_results.append((units_sold, fixed_costs, max_profitable_cac))

# Print scenario results
for result in scenario_results:
    print(f"For {result[0]} units sold and fixed costs of ${result[1]}:")
    print(f"Maximum CAC ensuring profitability: ${result[2]:.2f}\n")

# Save final analysis
scenario_df = pd.DataFrame(scenario_results, columns=["Units Sold", "Fixed Costs", "Max Profitable CAC"])
scenario_df.to_excel(file_path, sheet_name="Scenario Analysis", index=False)

print("Scenario analysis completed and saved successfully!")

# Visualization
plt.figure(figsize=(10, 5))
sns.lineplot(x=scenario_df["Units Sold"], y=scenario_df["Max Profitable CAC"], marker='o')
plt.xlabel("Units Sold")
plt.ylabel("Max Profitable CAC ($)")
plt.title("Impact of Sales Volume on Maximum Profitable CAC")
plt.grid(True)
plt.show()
```
